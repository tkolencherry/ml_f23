### Gradient Descent

#### Shape Vectors

```python
#for the p-vector we need to add a column of ones to represent x_0 
x_train = np.c_[np.ones(x_train.quarter.shape[0]), x_train]
x_train
```

#### Cost Function

```python
#generate random weights
np.random.seed(42)
weights = np.random.rand(6) # we need w0, w1, w2, w3, w4, w5, w6
weights
```

```python
def gradient(x,y,w):
        y_hat = np.dot(w,x.T) #y_hat = w0(1) + w1x1 + w2x2 +...
        residual = y_hat - y
        residual = residual.values
        x_0 = x[:, :1]
        x_1 = x[:, 1:2]
        x_2 = x[:, 2:3]
        x_3 = x[:, 3:4]
        x_4 = x[:, 4:5]
        x_5 = x[:, 5:6]
        #dot = np.dot(x_0.T,residual) - used for debugging
        return residual.mean(), np.dot(x_0.T,residual).mean(), np.dot(x_1.T,residual).mean(), np.dot(x_2.T,residual).mean(), np.dot(x_3.T,residual).mean(), np.dot(x_4.T,residual).mean(), np.dot(x_5.T,residual).mean()
        
```

```python
res, x0, x1, x2, x3, x4, x5 = gradient(x_train,y_train,weights)
res
```

```python
def gd(gradient, x, y, start, learn_rate = 0.1, n_iter = 10, tolerance = 1e-06):
    vector = start
    for _ in range(n_iter):
        diff = -learn_rate*np.array(gradient(x,y,vector)) 
        if np.all(abs(diff) != tolerance):
            break
        vector += diff[1:7] #the first column is the mean residuals, the rest are the partial derivatives w/ respect to x_i
    return vector
```

```python
sugg_weights = gd(gradient, x_train, y_train, weights, learn_rate = 0.1, n_iter = 10)
sugg_weights
```

```python
x_train[1]
```

```python
#now we need to test the linear regression line vs the actual data
train_yhat = []

#use our suggested weights to generate a y-hat array
for subset in x_train:
    temp_yhat = sugg_weights[0] + sugg_weights[1]*subset[1] + sugg_weights[2]*subset[2] + sugg_weights[3]*subset[3] + sugg_weights[4]*subset[4] + sugg_weights[5]*subset[5]
    train_yhat.append(temp_yhat)

#then calculate the training MSE
train_resid = (train_yhat-y_train).values
train_SE = train_resid**2
train_MSE = train_SE.mean()
train_MSE
```
